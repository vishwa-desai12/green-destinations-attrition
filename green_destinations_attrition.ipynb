```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set Seaborn style
sns.set(style="whitegrid")

# Load Dataset
df = pd.read_csv('greendestination.csv')  # Make sure the path is correct

# Step 1: Basic Data Info
print(" Dataset Shape:", df.shape)
print("Column Names:", df.columns.tolist())
print("\n First 5 Rows:\n", df.head())

# Step 2: Calculate Attrition Rate
attrition_counts = df['Attrition'].value_counts()
attrition_rate = (attrition_counts['Yes'] / len(df)) * 100
print(f"\nüìä Attrition Rate: {attrition_rate:.2f}%")

#  Step 3: Split into Two Groups
left = df[df['Attrition'] == 'Yes']
stayed = df[df['Attrition'] == 'No']

#  Step 4: Compare Key Factors
comparison = pd.DataFrame({
    'Average_Age': [left['Age'].mean(), stayed['Age'].mean()],
    'Average_YearsAtCompany': [left['YearsAtCompany'].mean(), stayed['YearsAtCompany'].mean()],
    'Average_MonthlyIncome': [left['MonthlyIncome'].mean(), stayed['MonthlyIncome'].mean()],
    'Average_DistanceFromHome': [left['DistanceFromHome'].mean(), stayed['DistanceFromHome'].mean()],
    'Average_JobSatisfaction': [left['JobSatisfaction'].mean(), stayed['JobSatisfaction'].mean()]
}, index=['Left', 'Stayed'])

print("\n Comparison of Features:\n", comparison)

# Step 5: Create Images Folder
os.makedirs("images", exist_ok=True)

#  Step 6: Visualizations

# 1. Attrition Count
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Attrition', palette='Set2', order=['Yes', 'No'])
plt.title('Attrition Count')
plt.ylabel('Number of Employees')
plt.savefig('images/attrition_count.png')
plt.show()

# 2. OverTime vs Attrition
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='OverTime', hue='Attrition', palette='Set1', order=['Yes', 'No'])
plt.title('Attrition by OverTime')
plt.ylabel('Number of Employees')
plt.savefig('images/overtime_attrition.png')
plt.show()

# 3. Age Distribution by Attrition
plt.figure(figsize=(8, 5))
sns.kdeplot(data=left, x='Age', label='Left', shade=True, color='red')
sns.kdeplot(data=stayed, x='Age', label='Stayed', shade=True, color='green')
plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.legend()
plt.savefig('images/age_distribution.png')
plt.show()

# 4. Monthly Income vs Attrition
plt.figure(figsize=(8, 5))
sns.kdeplot(data=left, x='MonthlyIncome', label='Left', shade=True, color='orange')
sns.kdeplot(data=stayed, x='MonthlyIncome', label='Stayed', shade=True, color='blue')
plt.title('Monthly Income Distribution by Attrition')
plt.xlabel('Monthly Income')
plt.legend()
plt.savefig('images/monthly_income_distribution.png')
plt.show()

print("\nAnalysis Complete! Charts saved in 'images/' folder.")
```
